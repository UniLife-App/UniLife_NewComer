```python
def getUserCate(dictInterestCate, dictUserInterest, dictUserCateCount):
    listInterests = dictUserInterest.keys()
    for interest in listInterests:
        if pd.isna(dictUserInterest[interest]) ==False:
            for cate in dictInterestCate[interest]:
                dictUserCateCount[cate] +=1
    return dictUserCateCount
```

```python
def getUserScore(dictUserCateCount, dictCateCount, dictUserCateScore):
    listCates = dictUserCateCount.keys()
    for cate in listCates :
        dictUserCateScore[cate] = round(dictUserCateCount[cate]/dictCateCount[cate]*7)
       
    return dictUserCateScore
```
